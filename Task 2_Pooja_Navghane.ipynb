## Import Libraries :

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

## Data Collection :

```python
df=pd.read_csv("iris.csv")
df.head() 
```

### Generate Summary Stastistics :

```python
df.describe()
```

### Create histogram and boxplots for numeric features :

###  Boxplot :

```python
print(df.dtypes)
```

```python
plt.figure(figsize=(10,3))
sns.boxplot(x='SepalLengthCm', y='SepalWidthCm',data=df,palette='winter')
```

```python
plt.figure(figsize=(12,3))
sns.boxplot(x='PetalLengthCm', y='PetalWidthCm',data=df,palette='winter')
```

### Histogram :

```python
df.hist(figsize=(10, 6), color='skyblue', edgecolor='black')
plt.suptitle("Histograms")
plt.tight_layout()
plt.show()
```

###  Pairplot (scatterplot matrix)

```python

sns.pairplot(df, diag_kind='hist', corner=True)
plt.suptitle("Pairplot of Iris Features", y=1.02)
plt.show()
plt.figure(figsize=(5, 3))
```

### # Correlation matrix heatmap

```python

plt.figure(figsize=(5, 3))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Iris Features")
plt.show()
```

### 1.Feature Relationships (from Correlation Matrix)
Strong positive correlations:

petal length vs petal width = 0.96

sepal length vs petal length = 0.87

Weak or no correlation:

sepal width has weak or negative correlation with others.

###  2. Anomalies
A few outliers in sepal width (from boxplots), likely from Setosa, which has a wider range.

Some overlap in Versicolor and Virginica petal features  these are harder to distinguish and could cause misclassification in ML models.
